# 5.1 Poblaciones y Muestras

En el análisis estadístico de datos, usamos datos de unas pocas muestras seleccionadas para sacar conclusiones sobre la población de la que se tomaron estas muestras. 

El diseño correcto del estudio debe garantizar que los datos de la muestra sean representativos de la población de la que se tomaron las muestras.

Dado que se desconocen los parámetros que caracterizan a la población, tenemos que usar muestras para obtener estimaciones de esos parámetros.

```python
from IPython.display import Image

Image(filename='Imagenes/5.1.png')
```

$\textbf{Población:}$ Incluye todos los elementos de un conjunto de datos.

$\textbf{Muestra:}$ Consiste en una o más observaciones de la población.

Se puede tomar más de una muestra de la misma población.

Al estimar un parámetro de una población, normalmente no podemos medir a todos los sujetos. Tenemos que limitarnos a investigar una muestra (esperemos representativa) tomada de este grupo. 

Con base en la estadística de la muestra, es decir, el valor correspondiente calculado a partir de los datos de la muestra, utilizamos la inferencia estadística para sacar conclusiones sobre el parámetro correspondiente en la población.

$\textbf{Parámetro:}$ Característica de una distribución que describe una población. Ej: la media o la desviación estándar de una distribución normal.

$\textbf{Estadístico:}$ Un valor numérico que representa una propiedad de una muestra aleatoria. Ej:
- El valor medio de los datos de la muestra.
- El rango de los datos de la muestra.
- Desviación de los datos de la media muestral.

$\textbf{Distribución muestral (empírica):}$ Distribución de probabilidad de un estadístico dado basado en una muestra aleatoria.

$\textbf{Inferencia estadística:}$ Le permite hacer una conjetura fundamentada sobre un parámetro de población en función de un estadístico calculado a partir de una muestra representativa de esa población.

Comparación de estadísticos muestrales y parámetros poblacionales

|           | Muestra estadístico | Parámetro de población |
|-----------|--------------------|-----------------------|
| Media     | $\bar{x}$          | $\mu$                 |
| Desviación estándar | $sd$         | $\sigma$              |

## 5.3 Distribuciones de probabilidad

### 5.3.1 Definiciones

#### a) Variables y variaciones aleatorias

Variable aleatoria $X$
- Es una cantidad que depende de eventos aleatorios. 
- Ej: el valor de un dado lanzado, puede tomar cualquier número del 1 al 6. 

Variación aleatoria
- Es un resultado particular o “realización” de una variable aleatoria. 
- Ej: lanzar un dado 3 veces y obtener los valores [1, 3, 4], esos valores serían 3 variaciones aleatorias.

#### b) Distribución de probabilidad

Describen la aleatoriedad de los datos en poblaciones y muestras.  

Para distribuciones

1) Discretas: se utilizan las Funciones de Masa de Probabilidad (PMF) 

* Al lanzar un dado, cada uno de los números del 1 al 6 tiene una probabilidad de 1/6 de salir. 
* Para todos los demás números, la probabilidad es cero. 
* Estos valores definen la PMF para lanzar dados. 
* El punto señalado indica que la probabilidad de sacar un 4 es 1/6.

```python
sns.set(font_scale=1.5)
sns.set_style('whitegrid')

pmf = np.zeros(10)
pmf[1:7] = 1/6

plt.plot(4, 1/6, 'o', ms=14, color='C1', alpha=0.5)
plt.plot(pmf, 'o')

# Anoto
xi, yi = 4, 1/6
xy_text = (5, 0.35)
plt.annotate('PMF(4)', xy = (xi, yi), xytext = xy_text, arrowprops=dict(facecolor='black', shrink=0.05) )

plt.axhline(y=0, ls='dotted')
plt.axvline(x=0, ls='dotted')

plt.xlabel('n')
plt.ylabel('PMF(n)')
plt.ylim(-0.02, 0.4)

plt.tight_layout()

plt.show()
```

2) Continuas: se utilizan las Funciones de Densidad de Probabilidad (PDF).

* Sea $p(x)$ la PDF de una variable aleatoria $X$. 
* La integral sobre p(x) entre a y b representa la probabilidad de encontrar el valor de $X$ en ese rango.

```python
# Datos
a, b = -0.5, 1.5 
x = np.linspace(-3, 3)
func = stats.norm().pdf
y = func(x)

fig, ax = plt.subplots()
plt.plot(x, y, 'r', linewidth=2)
plt.ylim(ymin=0)

# Hacer la región sombreada
ix = np.linspace(a, b)
iy = func(ix)
plt.fill_between(ix, iy, facecolor=[0.9, 0.9, 0.9], edgecolor=[0.5, 0.5, 0.5])

fs = 25
plt.text(0.5 * (a + b), 0.1, r"$\int_a^b p(x)\mathrm{d}x$", horizontalalignment='center', fontsize=fs)

plt.figtext(0.9, 0.05, '$x$', fontsize=fs)
plt.figtext(0.2, 0.95, '$p(x)$', fontsize=fs)
plt.axvline(x=-2, ymin=0, ymax=1, color='k', linewidth=1)

for loc in ['left', 'right', 'top']:
    ax.spines[loc].set_visible(False)

# Etiquete los bordes de la región sombreada
ax.xaxis.set_ticks_position('bottom')
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'), fontsize=fs)
ax.set_yticks([])

plt.show()
```

### 5.3.2 Distribuciones discretas

Ej: Lanzar dados.

Para cada uno de los números i = 1, ..., 6, la probabilidad de que al lanzar un dado, el lado que muestra el número i quede hacia arriba, $P_i$ , es 

$P_i = \frac{1}{6}, i = 1,...,6$.

```python
Image(filename='Imagenes/5.2.png')
```

La distribución de probabilidad para lanzar dados es el conjunto de todas estas probabilidades {$P_i$}. 

El valor más pequeño posible para $P_i$ es 0. 

Dado que una de las caras del dado tiene que aparecer en cada lanzamiento de dados, tenemos

$\sum\limits_{i=1}^{6} P_i = 1$

Propiedades:
- $0 \leq P_i \leq 1 \quad \forall i \in \mathbb{Z}$.
- $\sum\limits_{i} P_i = 1$ 

$P_i = PMF$

### 5.3.3 Distribuciones Continuas 

No está restringido a valores enteros discretos.

Ej: el peso de una persona puede ser cualquier número positivo. 

PDF:
- Es una función continua (por tramos) p(x).
- La curva describe la distribución de probabilidad, es decir, la probabilidad de cada valor.
- Es una función que describe la probabilidad relativa de que una variable aleatoria $X$ tome un valor cercano a x.
- Propiedades:
    - $p(x) \geq 0 \quad \forall x \in \mathbb{R}$.
    - La probabilidad de que ocurra un valor entre a y b es $p(a < x < b) = \int_{a}^{b} p(x) \, dx$
    - $\int_{-\infty}^{\infty} p(x) \, dx = 1$

### 5.3.4 Valor esperado y varianza

#### 5.3.4.1 Valor esperado

La PDF nos permite calcular el valor esperado E[X] o primer momento de una distribución continua de $X$:

$ E[X] = \int_{-\infty}^{\infty} x ~ p(x) \, dx $.

Se indica con $\mu$. 

Es la media de la población.

Para distribuciones discretas

$ E[X] = \sum\limits_{i} x_i P_i$

donde $x_i$ representa todos los posibles valores que puede tener la variable aleatoria. 

El valor esperado solo depende de la distribución de probabilidad.

```python
Image(filename='Imagenes/5.1.png')
```

La figura muestra un esquema de cómo se relaciona un estadístico de muestra con el parámetro de población correspondiente. 

La media muestral de nuestra muestra, $\bar{x}$, es el valor medio observado de la muestra. 

Si el experimento se diseñó correctamente, la incertidumbre esperada en la media de la muestra debería disminuir a medida que se incluyen más y más datos en la muestra.

#### 5.3.4.2 Varianza

También denominada segundo momento estandarizado.

Caracteriza la variabilidad de una variable aleatoria $X$:

$Var(X) = E[(X - E[X])^2] = E[X^2] - (E[X])^2$

Depende únicamente de la distribución de $X$, y se denota con $\sigma^2$, donde $\sigma$ es la desviación estándar de la población.

Cuando calculamos la varianza de la muestra, generalmente no conocemos el valor real de E[X] = $\mu$. 

Podemos reemplazar $\mu$ con el valor promedio de nuestro conjunto de datos, $\bar{x}$, minimizando así el error cuadrático medio. 

La varianza muestral esperada $\tilde{s}$, también llamada varianza muestral sesgada, es ligera pero sistemáticamente menor que la varianza poblacional:

$E[\tilde{s}^2] = \frac{n-1}{n} \sigma^2$,

donde n es el tamaño de la muestra. 

La mejor estimación de la varianza de la población es la varianza de la muestra (imparcial), denominada $s^2$:

$ s^2 = \frac{n}{n-1} \tilde{s}^2 = \frac{\sum\limits_i (x_i - \bar{x})^2}{n-1}$

También se llama varianza de muestra abreviada. 

Es lo que se usa comúnmente cuando se trabaja con datos.

#### 5.3.4.3 Deviación Estándar

Es la raíz cuadrada de la varianza.

Tiene la ventaja de que tiene la misma dimensión que $X$. 

La desviación estándar de la muestra es la raíz cuadrada de la varianza de la muestra:

$ s = \sqrt{\frac{\sum\limits_i (x_i - \bar{x})^2}{n-1}} $

Desviación estándar de la población: $\sigma$ 

Desviación estándar de la muestra: $s$

## 5.4 Grados de libertad

En mecánica, una partícula que se mueve en un plano tiene
- "2 DOF": en cada momento, 2 parámetros (coordenadas x/y) definen la ubicación de la partícula. 
- "3 DOF": si la partícula se mueve en el espacio, tiene  las coordenadas x/y/z. 

En estadística, un grupo de $n$ valores tiene $n DOF$. 

Si restamos la media de la muestra de cada valor, entonces, los datos restantes solo tienen $n-1 DOF$. 

Ej: n = 2: si conocemos el valor medio y el valor de $muestra_1$, entonces podemos calcular el valor de $muestra_2$ mediante $val_2 = 2 ∗ media − val_1$. 

Cuando tenemos muchos grupos, podemos usar el análisis de varianza (ANOVA).

Ej: 22 pacientes divididos en 3 grupos, los grados de libertad se dividen de la siguiente manera: 
- 1 DOF para el valor medio total. 
- 2 DOF para el valor medio de cada uno de los 3 grupos. 
    - Si conocemos los valores medios de 2 grupos y la media total, podemos calcular el valor medio del tercer grupo. 
- Quedan 19 DOF (= 22 − 1 − 2) para las desviaciones residuales de las medias del grupo.

## 5.5 Diseño del estudio

Sin un diseño de estudio riguroso, existe un sesgo significativo para obtener el resultado esperado.

### 5.5.1 Terminología

Factores o tratamientos: insumos controlados.

Cofactores, factores molestos o factores de confusión: entradas no controladas.

```python
Image(filename='Imagenes/5.4.png')
```

Covariable: variable que posiblemente predice el resultado que se está estudiando y puede ser un factor o un cofactor.

Cuando tratamos de modelar un proceso con 2 entradas y 1 salida, podemos formular un modelo matemático.

Ej:

$ Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_{12} X_1 X_2 + \epsilon $

Efectos principales: términos con una sola X ($\beta_1$, $\beta_2$) 

Términos de interacción: términos con múltiples X ($\beta_{12}$). 

Modelo lineal general: los parámetros $\beta$ entran en la ecuación solo de forma lineal.

Residuales($\epsilon$): se espera que se distribuyan aproximadamente normal alrededor de 0 si el modelo describe los datos correctamente.

Medida de resultado primaria: 
- Es el resultado que un investigador considera que es el más importante. 
- Debe definirse antes del comienzo de la recopilación de datos, para evitar que el investigador escoja resultados significativos y los presente como los hallazgos principales del estudio. 
- Forma la base para el cálculo del tamaño de muestra requerido.

### 5.5.2 Resumen

El primer paso en el diseño de un estudio es la aclaración explícita del objetivo del estudio. 

¿Queremos
1. comparar 2 o más grupos, o un grupo con un valor fijo?
    - Conduce a una prueba de hipótesis.
2. evaluar las respuestas observadas para identificar factores/efectos que son importantes? 
    - Es una investigación de cribado, en la que se debe estar atento a los artefactos si los factores del modelo son completamente independientes.
3. maximizar o minimizar una respuesta (variabilidad, distancia al objetivo, solidez)?
    - Es un problema de optimización.
4. desarrollar un modelo de regresión para cuantificar la dependencia de una variable de respuesta en la entrada del proceso?
    - Modelización estadística.

Una vez que hemos determinado lo que queremos hacer, tenemos que decidir cómo queremos hacerlo. 

Se pueden usar experimentos para obtener los datos
- Controlados
    - Intentamos variar solo un parámetro e investigar los efectos de ese parámetro en la salida.
- Observaciones 

### 5.5.3 Tipos de Estudios

#### a) Observacional o Experimental

Observacional:
- El investigador solo recopila información, pero no interactúa con la población de estudio.
- Se conocen como estudios no controlados.

Experimental:
- El investigador influye deliberadamente en los eventos e investiga los efectos de estas intervenciones. 
- Se conocen como estudios controlados.
- Ej: tratar al paciente con un nuevo tipo de medicamento. 

#### b) Prospectivo or Retrospectivo

Prospectivo: los datos se recopilan desde el comienzo del estudio. 

Retrospectivo: toma datos adquiridos de eventos anteriores
- Ej: pruebas de rutina realizadas en un hospital.

#### c) Longitudinal o Transversal

Longitudinal: el investigador recopila información durante un período de tiempo, tal vez varias veces, de cada paciente. 
- Ej: los experimentos suelen ser longitudinales.

Transversal: los individuos se observan una sola vez. 
- Ej: las encuestas suelen ser transversales. 

#### d) Estudios de Casos y Controles y de Cohortes

De casos y controles:
- Primero se trata a los pacientes y luego se los selecciona para incluirlos en el estudio, según ciertos criterios.
- Ej: si respondieron a un determinado medicamento. 

De cohorte:
- Los sujetos de interés se seleccionan primero y luego estos sujetos se estudian a lo largo del tiempo.
- Ej: por su respuesta a un tratamiento.

#### e) Ensayo controlado aleatorio

Se evita el sesgo dividiendo a los sujetos a evaluar en un grupo de intervención y un grupo de control. 

La asignación de grupos es aleatoria.

En un experimento diseñado, puede haber varias condiciones (factores) que son controladas por el experimentador. 

Al hacer que los grupos difieran en un solo aspecto (el factor de tratamiento) uno debería poder detectar el efecto del tratamiento en los pacientes.

A través de la aleatorización, los factores de confusión deben equilibrarse entre los grupos.

#### f) Estudios cruzados

Es un estudio longitudinal en el que los sujetos reciben una secuencia de diferentes tratamientos. 

Cada tema recibe cada tratamiento. 

El sujeto “cruza” de un tratamiento al siguiente. 

Para evitar efectos causales, la secuencia de asignación del tratamiento debe ser aleatoria.

Ej: investigación que prueba el efecto de estar de pie y sentado en la concentración de los sujetos, cada sujeto realiza tanto la ejecución de tareas de pie como la ejecución de tareas sentado. La secuencia de estar de pie/sentado es aleatoria, para cancelar cualquier efecto de secuencia.

### 5.5.4 Diseño de Experimentos

Bloquea todo lo que puedas; y aleatoriza el resto!

Tenemos factores (que podemos controlar) y cofactores (que no podemos controlar y/o manipular) que influyen en los resultados. 

Ej: tenemos un experimento en el que los resultados dependen de la persona que realiza el experimento (por ej, la enfermera que examina al sujeto) y de la hora del día. 
- Podemos bloquear el factor enfermera, haciendo que todas las pruebas sean realizadas por la misma enfermera. 
- No será posible evaluar todas las materias al mismo tiempo. 
- Tratamos de promediar los efectos del tiempo, mezclando aleatoriamente el tiempo de los sujetos. 
- Si medimos a nuestros pacientes por la mañana y a nuestros sujetos sanos por la tarde, invariablemente traeremos algún sesgo a nuestros datos.

#### a) Selección de la muestra

Al seleccionar los temas, se debe prestar atención a lo siguiente:

1. Las muestras deben ser representativas del grupo a estudiar. 
    - Ej: sujetos seleccionados al azar de pacientes en un hospital sesgan automáticamente la muestra hacia sujetos con problemas de salud.
    
2. En los estudios comparativos, los grupos deben ser similares con respecto a las fuentes conocidas de variación (por ej., edad, ...). 

3. ¡Asegúrese de que su selección de muestras (o temas) cubra suficientemente todos los parámetros que necesita!
    - Ej: si la edad es un factor molesto, asegúrese de tener suficientes sujetos jóvenes, de mediana edad y ancianos.
    - Ej: las pruebas de eficacia de una nueva terapia de rehabilitación para pacientes con ictus no solo deben incluir pacientes que hayan sufrido un ictus: asegúrese de que haya un número igual de pacientes con síntomas leves, medios y graves. 
        - De lo contrario, uno puede terminar con datos que incluyen principalmente pacientes con pocas o ninguna secuela del accidente cerebrovascular. 
        
El campo del “emparejamiento por puntajes de propensión” intenta corregir estos problemas.

#### b) Tamaño de la muestra

Muchos estudios fallan porque el tamaño de la muestra es demasiado pequeño para observar un efecto de la magnitud deseada. 

Para determinar el tamaño de la muestra, uno tiene que saber
- ¿Cuál es la varianza del parámetro bajo investigación? 
- ¿Cuál es la magnitud del efecto esperado, en relación con la desviación estándar del parámetro?

Esto se conoce como análisis de potencia. 

#### c) Sesgo

Un tamaño de muestra grande por sí solo no garantiza una respuesta representativa. Hay que tener cuidado con el sesgo de selección y el sesgo de no respuesta.

En general, al seleccionar los temas, se intenta que sean representativos del grupo a estudiar; y uno trata de conducir los experimentos de una manera representativa de las investigaciones de otros investigadores. Sin embargo, es muy fácil obtener datos sesgados.

El sesgo puede tener varias fuentes: 
- La selección de sujetos. 
- La estructura del experimento. 
- El dispositivo de medición. 
- El análisis de los datos. 

Se debe tener cuidado para evitar sesgos en los datos tanto como sea posible.

#### d) Aleatorización

Se usa para evitar el sesgo tanto como sea posible, y existen diferentes formas de aleatorizar un experimento. 

Se pueden utilizar generadores de números aleatorios. 
- Deben presentarse al experimentador lo más tarde posible.

Formas de aleatorizar la asignación del grupo:
- Aleatorización simple
    - Este procedimiento es robusto contra la selección y el sesgo accidental. 
    - Desventaja: los tamaños de los grupos resultantes pueden diferir significativamente. 

- Aleatorización por bloques
    - Se utiliza para mantener el número de sujetos en los diferentes grupos estrechamente equilibrado en todo momento. 
    - Ej: con 2 tipos de tratamiento, A y B, y un tamaño de bloque de 4, uno puede asignar los 2 tratamientos a los bloques de 4 sujetos en las siguientes secuencias
    1. AABB
    2. ABAB
    3. ABBA
    4. BBAA
    5. BABA
    6. BAAB
    - Se puede usar un generador de números aleatorios para generar enteros aleatorios entre 1 y 6, y usar los bloques correspondientes para asignar los respectivos tratamientos. 
    - Mantendrá el número de sujetos en cada grupo siempre casi igual.

- Minimización
    - Se toma el tratamiento que tenga el menor número de sujetos y se asigna este tratamiento con una probabilidad superior a 0.5 al próximo paciente. 
    - Ej: está realizando un ensayo controlado aleatorio de un nuevo medicamento, con un “grupo de placebo” y un “grupo de medicamento real”. 
    - A la mitad de los ensayos, se da cuenta de que su grupo de placebo ya contiene 60 sujetos, mientras que su grupo de medicación solo tiene 40. 
    - Ahora puede resolver este desequilibrio, dando a cada sujeto restante un 60% de probabilidad (en lugar del 50% utilizado anteriormente) el medicamento en lugar del placebo.

- Aleatorización estratificada 
    - Uno puede querer incluir una variedad más amplia de sujetos, con diferentes características. 
    - Ej: uno puede optar por tener sujetos tanto más jóvenes como mayores.
    - Se debe tratar de mantener equilibrado el número de sujetos dentro de cada estrato. 
    - Se deben mantener listas separadas de números aleatorios para cada grupo de sujetos.
    - Para muchos tipos de análisis de datos, es importante tener el mismo número de muestra en cada grupo. 

#### e) Cegamiento

El experimentador puede influir significativamente en el resultado de un experimento. 

Ej: un joven investigador con una nueva idea “brillante” para un nuevo tratamiento estará sesgado en la ejecución del experimento, así como en el análisis de los datos, para ver confirmada la hipótesis. 

Para evitar tal influencia subjetiva, idealmente tanto el experimentador como el sujeto deberían estar cegados a la terapia. 

Esto se conoce como doble ciego. 

Cuando además la persona que hace el análisis no sabe a qué grupo ha sido asignado el sujeto, hablamos de triple ciego.

#### f) Diseño factorial

Diseño factorial completo del experimento: cuando se prueba cada combinación de factores. 

Al planificar el análisis, se debe distinguir entre comparaciones
- Dentro de sujetos
    - Dentro de las comparaciones de sujetos, permite detectar diferencias más pequeñas con el mismo número de sujetos que entre comparaciones de sujetos.
- Entre sujetos. 

### 5.5.5 Recomendaciones para investigadores

#### Realizar un cálculo de potencia a priori

Utilice la literatura existente para estimar el tamaño del efecto que está buscando y diseñe su estudio en consecuencia. 

Si las limitaciones de tiempo o financieras significan que su estudio no tiene suficiente poder estadístico, aclare y reconozca esta limitación (o limitaciones) en la interpretación de sus resultados. 

#### Revelar métodos y hallazgos de manera transparente

Si los análisis previstos producen hallazgos nulos y pasa a explorar sus datos de otras maneras, dígalo. 

Los hallazgos nulos encerrados en cajones de archivos sesgan la literatura, mientras que los análisis exploratorios solo son útiles y válidos si reconoce las advertencias y limitaciones.

#### Pre-registro de su Protocolo de Estudio y Plan de Análisis

El prerregistro aclara si los análisis son confirmatorios o exploratorios, fomenta estudios bien fundamentados y reduce las oportunidades para la extracción de datos no transparentes y la elaboración de informes selectivos.

#### Hacer que los materiales de estudio y los datos estén disponibles

La disponibilidad de materiales de investigación mejorará la calidad de los estudios destinados a replicar y ampliar los resultados de la investigación. 

Hacer que los datos sin procesar estén disponibles mejorará las oportunidades para la agregación de datos y el metanálisis, y permitirá la verificación externa de los análisis y resultados.

#### Trabaje en colaboración para aumentar el poder y replicar los hallazgos

La combinación de datos aumenta el tamaño total de la muestra (y, por lo tanto, el poder) al tiempo que minimiza el impacto de la mano de obra y los recursos en cualquier contribuyente.

4. Almacena tus datos de forma bien estructurada.

#### Investigaciones Preliminares y la Ley de Murphy

La mayoría de las investigaciones requieren más de una ronda de experimentos y análisis. 

Teóricamente, estableces tu hipótesis primero, luego haces los experimentos y finalmente aceptas o rechazas la hipótesis. 

La mayoría de las investigaciones reales han sido menos sencillas y, por lo general, tomaron 2 rondas de experimentos. 

Por lo general, empiezo con una idea. Después de asegurarme de que nadie más ha encontrado la solución todavía, me siento, realizo las primeras rondas de mediciones y escribo los programas de análisis necesarios para analizar los datos. 

A través de esto, encuentro la mayoría de las cosas que pueden salir mal (normalmente lo hacen, como lo establece la Ley de Murphy: "Todo lo que puede salir mal, saldrá mal"), y lo que debería haber hecho de manera diferente en primer lugar. Si los experimentos tienen éxito, esa primera ronda de investigación me proporciona una "prueba de principio" de que mi pregunta es tratable; además, obtengo datos sobre la variabilidad de las respuestas típicas. Esto me permite obtener una estimación razonable del número de sujetos/muestras necesarios para aceptar o rechazar mi hipótesis. 

En este momento, también sé si mi configuración experimental es suficiente o si se requiere una configuración diferente o mejor. 

La segunda ronda de investigaciones es, en la mayoría de los casos, real y (si tengo suerte) me proporciona suficientes datos para publicar mis hallazgos.

#### Ejecuciones de calibración

Numerosos artefactos pueden influir en las mediciones de datos. 

Para controlar estos artefactos tanto como sea posible, siempre se debe comenzar y finalizar la grabación experimental con algo conocido. 

Tener una grabación con conocimiento exacto de lo que está sucediendo ayuda a 
- Detectar desviaciones en los sensores y problemas en la configuración experimental. 
- Verificar la precisión de los programas de análisis.

#### Documentación

Asegúrese de documentar todos los factores que pueden influir en sus resultados y todo lo que sucede durante el experimento:
- La fecha y hora del experimento. 
- Información sobre los experimentadores y los sujetos. 
- El paradigma exacto que ha elegido. 
- Cualquier cosa destacable que suceda durante el experimento.

Sea lo más breve posible, pero anote todo lo notable que suceda durante el experimento. 

Sea claro acerca de los nombres de los archivos de datos registrados, ya que serán lo primero que necesitará cuando analice los datos más adelante. 

A menudo, no necesitará todos los detalles de sus notas. Pero cuando tiene valores atípicos o puntos de datos inusuales, estas notas pueden ser invaluables para el análisis de datos.

#### Almacenamiento de datos

Trate de tener convenciones de nomenclatura claras, intuitivas y prácticas.

Almacene siempre los datos sin procesar inmediatamente, preferiblemente en un directorio separado. Prefiero que este directorio sea de solo lectura, de modo que no elimine inadvertidamente datos sin procesar valiosos. 

En la mayoría de los casos, puede rehacer fácilmente una ejecución de análisis. Pero a menudo, no podrá repetir un experimento.

### 5.5.7 Buen diseño de estudio: plan de investigación clínica

ISO 14155-1:2003. Obliga a la elaboración de un CIP, especificando
1. Tipo de estudio (p. ej., doble ciego y con o sin grupo de control). 
2. Discusión del grupo de control y el procedimiento de asignación. 
3. Descripción del paradigma.
4. Descripción y justificación del criterio principal de valoración del estudio. 
5. Descripción y justificación de la variable de medida elegida. 
6. Dispositivos de medida y su calibración. 
7. Criterios de inclusión de asignaturas. 
8. Criterios de exclusión de asignaturas. 
9. Punto de inclusión (“¿Cuándo un sujeto es parte del estudio?”). 
10. Descripción del procedimiento de medición. 
11. Criterios y procedimientos para las asignaturas que abandonan. 
12. Número de muestra elegido y nivel de significancia, y su justificación. 
13. Procedimiento para la documentación de efectos negativos o colaterales. 
14. Lista de factores que pueden influir en los resultados de la medición o en su interpretación. 
15. Procedimiento para la documentación, también para datos faltantes. 
16. Procedimiento de análisis estadístico. 
17. La designación de un monitor para la investigación. 
18. La designación de un investigador clínico. 
19. Especificaciones para el manejo de datos.
