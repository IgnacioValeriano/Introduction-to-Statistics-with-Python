```python
# Establecer los parámetros para una onda sinusoidal

rate = 50

freq = 2

duration  = 4

amp = 5

noise_amp = 0.8
```

```python
# Calcular los valores de los senos

delta = np.deg2rad(15)

dt = 1 / rate

omega = 2 * np.pi * freq
```

```python
t = np.arange(0, duration, dt)

data = amp * np.sin(omega*t + delta) + noise_amp * np.random.randn(len(t))
```

```python
df = pd.DataFrame({'T': t, 'Valores': data})

df.head()
```

```python
# Mostrar cómo generar una cadena formateada en Python

print(f'La primera muestra de tiempo es {t[0]:5.3f}, y el primer valor de datos es {data[0]:5.3f}')
```

```python
def save_txt(df: pd.DataFrame, out_file='data.csv') -> None:
    """ Guardar datos en formato ASCII

    Parámetros
    ----------
    df : Datos de entrada
    out_file : Archivo de salida; otros dos archivos ASCII con el mismo tallo también se generan

    """

    # Guarda los datos en formato CSV, lo que significa que por defecto
    # - Separados por una coma
    # - Con un nombre de columna
    # - Con un índice corriente en el lado izquierdo
    df.to_csv(out_file)

    # ¡Infórmele siempre al usuario cuando genere un archivo nuevo!
    # Si usa Python> 3.7, puede usar las "cadenas de formato"
    print(f'Los datos se han guardado en formato CSV para {out_file}')

    # Para versiones anteriores de Python tienes que usar
    # print(f'CSV-Data han sido guardados en {out_file}')

    # Archivo simple, separado por tabuladores, sin encabezado, sin índice
    simple_file = out_file.replace('.csv', '_tab.txt')
    df.to_csv(simple_file, sep='\t', header=False, index=False)
    print(f'Los datos se han guardado en {simple_file}')

    # Mostrar cómo agregar un encabezado de archivo a un archivo de texto existente
    with open(out_file, 'r') as original:
        text = original.read()

    modified_file = out_file.replace('.csv', '_modified.txt')
    with open(modified_file, 'w') as modified:
        modified.write('Este archivo fue generado el Sept 19\n')
        modified.write(f'Tasa de muestreo: {rate} [Hz]\n')
        modified.write(text)
    print(f'Se ha agregado un encabezado de archivo a {out_file}, y el nuevo archivo guardado como {modified_file}')
```

```python
save_txt(df)
```

```python
def save_xls(df: pd.DataFrame, out_file='data.xlsx') -> None:
    """ Guardar datos en formato MS Excel

    Parámetros
    ----------
    df : Datos de entrada
    out_file : Nombre del archivo de salida
    """

    # Guardar datos en formato Excel
    
    df.to_excel(out_file, index=False)
    
    print(f'Los datos se han guardado en formato MS-Excel, para {out_file}')
```

```python
save_xls(df)
```

```python
def save_matlab(t: np.ndarray, data: np.ndarray, out_file='data.mat') -> None:
    """ Guardar datos en formato Matlab

    Parámetros
    ----------
    t :Valores de tiempo [sec]
    data : onda sinusoidal
    out_file : Nombre del archivo de salida
    """
    # Para guardar datos en formato Matlab, necesitamos el paquete "scipy.io"
    from scipy.io import savemat

    # y tenemos que poner los datos en un diccionario de Python
    # Para este ejemplo, agrego un texto de información y formateo los datos como matriz
    
    data_mat = np.column_stack( (t, data) )
    
    data_dict = {'info':'Estos son datos de demostración, que muestran una onda sinusoidal',
                'data': data_mat}

    savemat(out_file, data_dict)
    print(f'Los datos se han guardado en formato Matlab, para {out_file}')
```

```python
save_matlab(t, data)
```

```python
def generate_binary(out_file='data.raw') -> None:
    """Genere datos binarios, con un encabezado ASCII de 256 bytes y tres columnas de datos.

    Parámetros
    ----------
    out_file : nombre del archivo externo
    """

    # Para generar una señal más interesante, produzco un "chirrido"
    from scipy.signal import chirp

    # Establecer los parámetros
    length_header = 256     # byte

    # Generar un texto de encabezado ficticio
    txt = """Este es el encabezado.

    Tiene una longitud de byte 'length_header'. Después del texto, se rellena con espacios en blanco."""
    out_txt = txt + ' '*(length_header-len(txt))

    # Escribirlo en el archivo out_file
    fh = open(out_file, 'wb')
    fh.write(out_txt.encode())

    # Generar algunos datos
    t = np.arange(0,20, 0.1)
    x = np.sin(t)
    y = chirp(t, 3, np.max(t), 0.01)
    data = np.column_stack((t, np.sin(t), y*t))

    # También escríbalos en un archivo
    fh.write(data.tobytes())
    fh.close()
    print(f'Los datos se han escrito en forma binaria para {out_file}')
```

```python
generate_binary()
```

$\textbf{Ejercicio 2:}$ Modificación de archivos de texto: Números imaginarios.

El archivo imaginary.txt contiene las partes real e imaginaria de los números complejos, incluido un encabezado para cada columna. 

Escriba un script de Python que lea esos datos y agregue la representación polar de cada punto de datos (radio/ángulo [rad]) a cada línea, con el nombre del archivo de salida imaginary_out.txt.

Sugerencia: para el archivo de salida resultante, separe las etiquetas de encabezado y los números con una simple pestaña.

```python
import os
```

```python
# Establecer los parámetros requeridos

data_dir = 'Datos'

file_name = 'data_tab.txt'

in_file = os.path.join(data_dir, file_name)

out_file = 'imaginary_out.txt'
```

```python
# Obtener los datos

df = pd.read_csv(in_file, delim_whitespace = True, header = None)

df.head()
```

```python
# Asigna nombres de columnas específicos

df.columns = ['Real', 'Imaginary']
```

```python
df.head()
```

```python
# Agregar radio y ángulo como nuevas columnas

df['Radius'] = np.sqrt(df.Real**2 + df.Imaginary**2)

df['Angle [rad]'] = np.arctan2(df.Imaginary, df.Real)
```

```python
df.head()
```

```python
# Asegúrese de que todas las columnas sean flotantes y escriba en el nuevo archivo

df = df.astype('float')

df.to_csv(out_file, sep = '\t', index = None, float_format = '%5.3f')

print(f'Datos modificados guardados en {out_file}')
```

## Pickle

Se puede usar el módulo "pickle" para guardar un objeto de Python para su uso posterior. 

Por lo tanto, es conveniente poner todos los datos requeridos en un diccionario:

```python
# Datos

t = np.arange(0, 10, 0.1)
x = np.sin(t)
info = 'Esta es una onda sinusoidal.'

data = {'t': t,
        'x': x,
        'info': info}
```

### 1. Guardando datos

La forma más conveniente de manejar objetos encurtidos es con la sintaxis provista por "with":

```python
import pickle
```

```python
# Guardarlos como un archivo pickled 

data_file = 'data.pickle'

with open(data_file, 'wb') as fh_out:
    pickle.dump(data, fh_out)
```

### 2. Cargando datos

```python
# Cargando 

with open(data_file, 'rb') as fh_in:
    loaded_data = pickle.load(fh_in)
```
